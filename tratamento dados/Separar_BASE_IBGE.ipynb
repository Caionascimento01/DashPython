```python
import geopandas as gpd

shapefile_path = '..\..\BR_Municipios_2024\BR_Municipios_2024.shp'

# Carrega o shapefile usando geopandas
gdf_municipios = gpd.read_file(shapefile_path)

shapefile_path = '..\..\BR_UF_2024\BR_UF_2024.shp'

# Carrega o shapefile usando geopandas
gdf_estados = gpd.read_file(shapefile_path)
```

```python
gdf_municipios.info()
```

```python
col = ['CD_MUN',
    'NM_MUN',
    'CD_UF',
    'NM_UF',
    'SIGLA_UF',
    'CD_REGIA',
    'NM_REGIA',
    'AREA_KM2',
    'geometry']

gdf_municipios = gdf_municipios[col]
gdf_municipios = gdf_municipios.rename(columns={'geometry': 'POLYGON'})    

gdf_municipios.head()
```

```python
gdf_estados = gdf_estados.rename(columns={'geometry': 'POLYGON'})  

gdf_estados.head()
```

```python
gdf_municipios.to_csv('..\datasets\gdf_municipios.csv', index=False)
gdf_estados.to_csv('..\datasets\gdf_estados.csv', index=False)
```

```python
merged_gdf = gdf_municipios.merge(gdf_estados, left_on='CD_UF', right_on='CD_UF', suffixes=('_municipio', '_estado'))

merged_gdf.head()
```
