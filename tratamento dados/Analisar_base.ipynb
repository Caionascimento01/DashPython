```python
import pandas as pd
```

```python
df = pd.read_csv('..\datasets\RECLAMEAQUI_CARREFUOR.csv', sep=',')

df.head()
```

```python
df['MUNICIPIO'] = df['LOCAL'].str.split(' - ').str[0]
df['SIGLA_UF'] = df['LOCAL'].str.split(' - ').str[1]

df.groupby(['SIGLA_UF']).size().reset_index(name='Qtd_Reclamacoes')
```

```python
df['TEMPO'] = pd.to_datetime(df['TEMPO']).dt.strftime('%d-%m-%Y')

df.head()
```

```python
df['STATUS'].value_counts()
```

```python
# Split the 'LOCAL' column into 'MUNICIPIO' and 'SIGLA_UF'
df['MUNICIPIO'] = df['LOCAL'].str.split(' - ').str[0]
df['SIGLA_UF'] = df['LOCAL'].str.split(' - ').str[1]

df.head()
```

```python
df[df['SIGLA_UF'] == '--'].head()
```

```python
# Remove rows where 'SIGLA_UF' is '--'
df = df[df['MUNICIPIO'] != 'naoconsta']
df['MUNICIPIO'].value_counts()
```

```python
df.head()
```

```python
# adicionar coluna com o nome do estado
df['NOME_UF'] = df['SIGLA_UF'].map({
    'AC': 'Acre',
    'AL': 'Alagoas',
    'AP': 'Amapá',
    'AM': 'Amazonas',
    'BA': 'Bahia',
    'CE': 'Ceará',
    'DF': 'Distrito Federal',
    'ES': 'Espírito Santo',
    'GO': 'Goiás',
    'MA': 'Maranhão',
    'MT': 'Mato Grosso',
    'MS': 'Mato Grosso do Sul',
    'MG': 'Minas Gerais',
    'PA': 'Pará',
    'PB': 'Paraíba',
    'PR': 'Paraná',
    'PE': 'Pernambuco',
    'PI': 'Piauí',
    'RJ': 'Rio de Janeiro',
    'RN': 'Rio Grande do Norte',
    'RS': 'Rio Grande do Sul',
    'RO': 'Rondônia',
    'RR': 'Roraima',
    'SC': 'Santa Catarina',
    'SP': 'São Paulo',
    'SE': 'Sergipe',
    'TO': 'Tocantins'
})

df.head()
```

```python
# remonear nome de coluna
df.rename(columns={'TRIMETRES': 'TRIMESTRE'}, inplace=True)

# Organizar as colunas
df = df[['ID', 'TEMPO', 'NOME_UF', 'SIGLA_UF', 'MUNICIPIO','STATUS', 'TEMA', 'CATEGORIA', 'DESCRICAO', 'DIA', 'MES', 'ANO', 'TRIMESTRE']]

df.head()
```

```python
df.to_csv('..\datasets\RECLAMEAQUI_CARREFUOR_CLS.csv', sep=',', encoding='utf-8')
```

```python
def count_palavras(texto):
    return len(texto.split())

df['DESCRICAO'].apply(count_palavras).plot(kind='hist', density=True)
df['DESCRICAO'].apply(count_palavras).plot.kde()
```

```python
# !pip install wordcloud

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud # https://amueller.github.io/word_cloud/index.html

# Obter a lista de stopwords em português
stopwords_portugues = stopwords.words('portuguese')

novas_stopwords = ["produto", "empresa", "comprei", "loja", "não", "pra", "tive", "minha", 
    "dia", "dias", "entrega", "reclame", "aqui", "problema", "já", "pois", "outro", "outra", "Ibyte"]

for palavra in novas_stopwords:
    stopwords_portugues.append(palavra)
```

```python
texto = " ".join(df['DESCRICAO'].astype(str).tolist())
print(len(texto)) 
print(texto[:100])  

# Gerar a nuvem de palavras
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    stopwords=stopwords_portugues,
    colormap='viridis', 
    max_words=50
).generate(texto)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud - Termos mais frequentes nas reclamações')
plt.show()
```
